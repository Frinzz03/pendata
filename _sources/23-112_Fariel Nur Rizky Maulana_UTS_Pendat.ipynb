# **Memahami data**
## Melakukan Proses pengumpulan data

```python
import pandas as pd
from tabulate import tabulate

# Ganti "data_obat.csv" sesuai nama file kamu
df = pd.read_csv("cirrhosis.csv")

# Tampilkan dengan tabulate
print(tabulate(df, headers='keys', tablefmt='grid'))
```

## Melakukan Sampling Data

```python
# Ambil sampel 10% data secara acak
sample_df = df.sample(frac=0.1, random_state=42)
print("Sample Data (10%):")
print(tabulate(sample_df.head(10), headers='keys', tablefmt='grid'))
```

## Melakukan profilling data



```python
from tabulate import tabulate

# Ringkasan data: tipe data dan jumlah nilai unik
print(" Tipe Data dan Jumlah Nilai Unik:")
print(tabulate(df.dtypes.reset_index().rename(columns={"index": "Kolom", 0: "Tipe Data"}), headers="keys", tablefmt="grid"))

print("\n Nilai Unik per Kolom:")
print(tabulate(df.nunique().reset_index().rename(columns={"index": "Kolom", 0: "Jumlah Unik"}), headers="keys", tablefmt="grid"))

# Statistik deskriptif untuk kolom numerik dan kategorikal
print("\n Statistik Deskriptif (Numerik):")
print(tabulate(df.describe().T, headers="keys", tablefmt="grid"))

print("\n Statistik Deskriptif (Kategorikal):")
print(tabulate(df.describe(include=['object']).T, headers="keys", tablefmt="grid"))
```

## Menentukan korelasi dan asosiasi

```python
# Korelasi Pearson antar variabel numerik
print("\n Korelasi Pearson antar Variabel Numerik:")
print(tabulate(df.corr(numeric_only=True), headers="keys", tablefmt="grid"))

# Contoh asosiasi: tabulasi silang (cross-tab) antar dua kolom kategorikal
print("\n Contoh Asosiasi Variabel Kategorikal (Sex vs Drug):")
if "Sex" in df.columns and "Drug" in df.columns:
    crosstab = pd.crosstab(df["Sex"], df["Drug"])
    print(tabulate(crosstab, headers='keys', tablefmt='grid'))
else:
    print("Kolom 'Sex' atau 'Drug' tidak ditemukan di dataset.")
```

## Melakukan Explorasi Data

```python
import pandas as pd
import scipy.stats as stats

# Pastikan combined_df sudah tersedia sebelum menjalankan kode ini
def explore_data(df):
    stats_dict = {
        "Mean": df.mean(),
        "Median": df.median(),
        "Mode": df.mode().iloc[0],  # Ambil nilai pertama jika ada beberapa modus
        "Std Dev": df.std(),
        "Min": df.min(),
        "Max": df.max()
    }
    return pd.DataFrame(stats_dict)

# Pastikan hanya kolom numerik yang dianalisis
numeric_columns = df.select_dtypes(include=['number'])

# Panggil fungsi eksplorasi data
stats_df = explore_data(numeric_columns)

# Tampilkan hasil eksplorasi data
print("\nStatistik Deskriptif:")
print(tabulate(stats_df, headers='keys', tablefmt='grid'))
```

## Mengidentifikasi Masalah

```python
# Missing Values
print("\n Missing Values:")
missing = df.isnull().sum()
missing = missing[missing > 0].reset_index()
missing.columns = ["Kolom", "Jumlah Missing"]
print(tabulate(missing, headers='keys', tablefmt='grid'))

# Duplikasi
print("\n Duplikasi Baris:")
duplicates = df.duplicated().sum()
print(f"Jumlah baris duplikat: {duplicates}")

# Outliers: Cek nilai ekstrem pada kolom numerik
print("\n Deteksi Outliers (Z-score sederhana pada beberapa kolom):")
from scipy.stats import zscore
z_scores = df.select_dtypes(include='number').apply(zscore).abs()
outliers = (z_scores > 3).sum().reset_index()
outliers.columns = ["Kolom", "Jumlah Outliers (Z>3)"]
print(tabulate(outliers[outliers["Jumlah Outliers (Z>3)"] > 0], headers='keys', tablefmt='grid'))

# Inkonsistensi contoh: Nilai kolom 'Sex' harus konsisten
print("\n Nilai Unik (cek inkonsistensi):")
for col in df.select_dtypes(include='object').columns:
    print(f"- Kolom '{col}': {df[col].unique()}")
```

# **Evaluasi**
Tujuan dari proses *data profiling* yang dilakukan adalah untuk mendapatkan pemahaman mendalam tentang struktur, kualitas, dan potensi informasi yang terkandung dalam dataset. Berikut penjelasan singkat evaluasi dan tujuannya:

1. **Tipe Data & Nilai Unik**  
   - **Evaluasi**: Mengidentifikasi tipe data (numerik/kategorikal) dan menghitung jumlah nilai unik di setiap kolom.
   - **Tujuan**: Menentukan jenis analisis atau transformasi yang sesuai, serta mengidentifikasi fitur yang mungkin kurang informatif atau perlu dibersihkan.

2. **Statistik Deskriptif**  
   - **Evaluasi**: Menghasilkan ringkasan statistik untuk kolom numerik (seperti mean, min, max) dan kategorikal (frekuensi terbanyak).
   - **Tujuan**: Memberikan gambaran umum distribusi data dan mendeteksi adanya nilai ekstrem atau ketidakseimbangan kategori.

3. **Korelasi & Asosiasi**  
   - **Evaluasi**: Mengukur kekuatan hubungan antar fitur numerik (korelasi Pearson) dan antar fitur kategorikal (tabulasi silang).
   - **Tujuan**: Mengidentifikasi fitur-fitur yang saling terkait untuk membantu pemilihan fitur dalam model dan mencegah multikolinearitas.

4. **Masalah Data**  
   - **Evaluasi**:
     - *Missing values*: Mengidentifikasi nilai kosong atau tidak tercatat.
     - *Duplikasi*: Menemukan baris data yang identik.
     - *Outliers*: Mendeteksi nilai ekstrem dengan Z-score.
     - *Inkonsistensi*: Menemukan nilai yang tidak seragam pada kolom kategorikal.
   - **Tujuan**: Menjamin kualitas data dengan memperbaiki atau menghapus elemen-elemen yang dapat mengganggu proses analisis atau pemodelan.

 Secara keseluruhan, kegiatan ini dilakukan untuk memastikan data bersih, relevan, dan siap digunakan dalam analisis lanjutan atau model prediktif, sehingga hasilnya lebih akurat dan terpercaya.
