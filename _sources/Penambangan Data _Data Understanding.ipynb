#Sampling Data
Sampling data adalah teknik dalam statistik untuk memilih sebagian kecil data dari populasi yang lebih besar guna menganalisis atau membuat prediksi tanpa harus menggunakan seluruh data. Sampling digunakan untuk menghemat waktu, biaya, dan sumber daya, sambil tetap memberikan hasil yang representatif.

#Identifikasi Masalah Data
Identifikasi masalah data adalah proses menemukan dan memahami isu-isu yang dapat mempengaruhi kualitas, keakuratan, dan kegunaan data dalam analisis atau pengambilan keputusan. Masalah data dapat menyebabkan hasil yang tidak valid, kesalahan dalam model, atau pengambilan keputusan yang salah.

1. Data Hilang (Missing Data)
2. Data Ganda (Duplicate Data)
3. Data Tidak Konsisten
4. Data Outlier (Nilai Ekstrem)
5. Data Tidak Valid
6. Data Redundansi
7. Data Skewed (Distribusi Tidak Merata)

#Profiling Data
Profiling data adalah proses menganalisis, memeriksa, dan memahami karakteristik suatu dataset sebelum digunakan dalam analisis lebih lanjut atau model pembelajaran mesin. Tujuannya adalah untuk mengevaluasi kualitas data, mendeteksi anomali, serta memastikan kelengkapan dan konsistensi data.

```python
import pymysql
import psycopg2
import pandas as pd

# Konfigurasi koneksi ke database Aiven MySQL
mysql_config = {
    "host": "mysql23-112-irismysql23-112.h.aivencloud.com",
    "user": "avnadmin",
    "password": "AVNS_cTNd4CUt1kk0Z7MPIC1",
    "database": "defaultdb",
    "port": 15786
}

# Konfigurasi koneksi ke database Aiven PostgreSQL
postgres_config = {
    "host": "postgre23-112-postgresql23-112.h.aivencloud.com",
    "database": "defaultdb",
    "user": "avnadmin",
    "password": "AVNS_Ellqk9Jx9uICNV-C29t",
    "port": 15184
}

# Fungsi untuk mengambil data dari MySQL (hanya petal length & petal width)
def fetch_mysql_data():
    try:
        conn = pymysql.connect(**mysql_config)
        cursor = conn.cursor()
        query = 'SELECT id, class, `petal length`, `petal width` FROM mysqliris.flowers;'
        cursor.execute(query)
        data = cursor.fetchall()
        columns = [desc[0] for desc in cursor.description]  # Ambil nama kolom
        cursor.close()
        conn.close()
        df = pd.DataFrame(data, columns=columns)

        # Ubah nama kolom agar seragam
        df = df.rename(columns={"petal length": "petal_length", "petal width": "petal_width"})

        return df
    except pymysql.err.OperationalError as e:
        print(f"Error connecting to MySQL: {e}")
        return pd.DataFrame()

# Fungsi untuk mengambil data dari PostgreSQL (hanya sepal length & sepal width)
def fetch_postgres_data():
    try:
        conn = psycopg2.connect(**postgres_config)
        cursor = conn.cursor()
        query = 'SELECT id, "Class", "sepal length", "sepal width" FROM postgresqliris.postgresql;'
        cursor.execute(query)
        data = cursor.fetchall()
        columns = [desc[0] for desc in cursor.description]  # Ambil nama kolom
        cursor.close()
        conn.close()
        df = pd.DataFrame(data, columns=columns)

        # Ubah nama kolom agar seragam
        df = df.rename(columns={"Class": "class", "sepal length": "sepal_length", "sepal width": "sepal_width"})

        return df
    except psycopg2.OperationalError as e:
        print(f"Error connecting to PostgreSQL: {e}")
        return pd.DataFrame()

# Ambil data dari kedua database
mysql_df = fetch_mysql_data()
postgres_df = fetch_postgres_data()

# Gabungkan data berdasarkan id dan class
combined_df = pd.merge(postgres_df, mysql_df, on=["id", "class"], how="inner")

# Tampilkan hasil
print("\nData Gabungan dari MySQL dan PostgreSQL:")
print(combined_df.to_string(index=False))
```

#Eksplorasi Data
Eksplorasi data adalah proses memahami dan menganalisis dataset sebelum melakukan pemodelan atau pengambilan keputusan. Tujuan utamanya adalah mengevaluasi struktur, distribusi, pola, serta kemungkinan anomali dalam data.

```python
import pandas as pd
import scipy.stats as stats

# Pastikan combined_df sudah tersedia sebelum menjalankan kode ini
def explore_data(df):
    stats_dict = {
        "Mean": df.mean(),
        "Median": df.median(),
        "Mode": df.mode().iloc[0],  # Ambil nilai pertama jika ada beberapa modus
        "Std Dev": df.std(),
        "Min": df.min(),
        "Max": df.max()
    }
    return pd.DataFrame(stats_dict)

# Pastikan hanya kolom numerik yang dianalisis
numeric_columns = combined_df.select_dtypes(include=['number'])

# Panggil fungsi eksplorasi data
stats_df = explore_data(numeric_columns)

# Tampilkan hasil eksplorasi data
print("\nStatistik Deskriptif:")
print(stats_df)
```

#Visualisasi Data
Visualisasi data adalah teknik untuk menyajikan data dalam bentuk grafik atau gambar agar lebih mudah dipahami. Tujuannya adalah untuk menemukan pola, tren, atau hubungan dalam data yang sulit terlihat hanya dengan angka.

##Visualisasi Distrbusi

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pastikan dataframe sudah tersedia
if 'combined_df' not in locals():
    print("Data belum tersedia. Harap jalankan proses pengambilan data terlebih dahulu.")
else:
    # Menentukan warna berdasarkan class
    palette = {"Iris-setosa": "blue", "Iris-versicolor": "orange", "Iris-virginica": "green"}

    #  Visualisasi Distribusi (Histogram)
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    sns.histplot(data=combined_df, x="sepal_length", hue="class", kde=True, ax=axes[0, 0], palette=palette)
    sns.histplot(data=combined_df, x="sepal_width", hue="class", kde=True, ax=axes[0, 1], palette=palette)
    sns.histplot(data=combined_df, x="petal_length", hue="class", kde=True, ax=axes[1, 0], palette=palette)
    sns.histplot(data=combined_df, x="petal_width", hue="class", kde=True, ax=axes[1, 1], palette=palette)

    axes[0, 0].set_title("Distribusi Sepal Length Berdasarkan Class")
    axes[0, 1].set_title("Distribusi Sepal Width Berdasarkan Class")
    axes[1, 0].set_title("Distribusi Petal Length Berdasarkan Class")
    axes[1, 1].set_title("Distribusi Petal Width Berdasarkan Class")

    plt.tight_layout()
    plt.show()
```
